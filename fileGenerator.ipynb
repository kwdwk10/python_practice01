```python
#ファイル書き出し
for j in range(3):
    filepath=str(j+1).zfill(4)
    print(filepath)
    os.makedirs(filepath,exist_ok = True)
    for i in range(10):
        tmp = pd.DataFrame({'a':np.repeat(i*(j+1),10),'b':np.repeat(i*(j+1),10),'c':np.repeat(i,10)})
        tmp.to_csv('./'+ filepath +'/'+ str(i)+'.csv',index=False)
```
